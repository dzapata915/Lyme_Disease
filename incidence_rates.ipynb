```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
count_by_state =pd.read_csv('../data/Case Count by State 2008-2023.csv')
```

```python
count_by_state
```

```python
count_by_state['State']=count_by_state['State'].str.replace('†','',regex=False)
count_by_state
```

```python
statecount_2023=count_by_state[['State','2023']]
```

```python
statecount_2023['2023'] = pd.to_numeric(statecount_2023['2023'].str.replace(',', '', regex=False), errors='coerce')
```

```python
statecount_2023.info()
```

```python
statecount_2023.head()
```

```python
state_pops=pd.read_csv('../data/Census Population by State 2020-2024.csv')
```

```python
state_pops
```

```python
state_pops.columns
```

```python
state_pops=state_pops[['Geographic Area',
       'Population Estimate (as of July 1)', 'Unnamed: 3', 'Unnamed: 4',
       'Unnamed: 5', 'Unnamed: 6']]
```

```python
state_pops=state_pops.drop(index=[2,3,4,5])
```

```python
state_pops.columns=['Geographic Area','Pop_2020','Pop_2021','Pop_2022','Pop_2023','Pop_2024']
```

```python
state_pops=state_pops.drop(index=0)
```

```python
state_pops['Geographic Area']=state_pops['Geographic Area'].str.replace('.','',regex=False)
```

```python
pops_2023= state_pops[['Geographic Area','Pop_2023']]
```

```python
pops_2023.head()
```

```python
statecount_2023.head()
```

```python
state_incidence= pd.merge(statecount_2023,pops_2023, left_on='State',right_on='Geographic Area')
```

```python
state_incidence=state_incidence.drop(columns='Geographic Area')
```

```python
state_incidence.columns=['State','Lyme_Case_Count_2023','Pop_2023']
```

```python
state_incidence.info()
```

```python
state_incidence['Pop_2023']=state_incidence['Pop_2023'].str.replace(',', '').astype(int)
```

```python
state_incidence.info()
```

```python
state_incidence['incidence_rate_per_100k']=round((state_incidence['Lyme_Case_Count_2023']/state_incidence['Pop_2023'])*100000,1)
```

```python
state_incidence.sort_values('incidence_rate_per_100k',ascending=False).head(10)
```

```python
state_incidence_sorted = state_incidence.sort_values(by='incidence_rate_per_100k', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=state_incidence_sorted,
    y='State',
    x='incidence_rate_per_100k',
    palette='viridis'
)

plt.title('Lyme Disease Incidence Rate 2023', fontsize=16)
plt.xlabel('Incidence Rate per 100,000', fontsize=14)
plt.ylabel('State', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
