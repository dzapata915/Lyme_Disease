```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
overall_LDcase_count = pd.read_csv('../../data/Overall US Case Count by Year 1996-2023.csv')
```

```python
overall_LDcase_count
```

```python
overall_LDcase_count.plot(kind='line',x='Year',y='Cases')
```

```python
plt.figure(figsize=(12, 6))

sns.lineplot(
    data=overall_LDcase_count,
    x='Year',
    y='Cases',
    marker='o',
    color='#556B2F',
    linewidth=2,
    markersize=6
)

key_years = [2008, 2011, 2017, 2020, 2022]
label_offset = 0.4  

for year in key_years:
    plt.axvline(x=year, color='gray', linestyle='--', linewidth=1)
    plt.text(year + label_offset, max(overall_LDcase_count['Cases']) * 0.02, str(year),
             rotation=0,
             verticalalignment='bottom',
             horizontalalignment='left',  
             fontsize=10,
             color='gray')

plt.title('Total Reported Lyme Disease Cases in the U.S. (19962023)', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Reported Cases', fontsize=12)
plt.ylim(bottom=0)
plt.tight_layout()
plt.show()
```

```python
cases_first_year=states_years_long.loc[states_years_long['Year'] == 2008][['State', 'Cases']]
```

```python
cases_last_year = states_years_long[states_years_long['Year'] == 2023][['State', 'Cases']]
```

```python
perc_change = pd.merge(cases_first_year, cases_last_year, on='State', suffixes=('_2008', '_2023'))
```

```python
perc_change['Pct_Change'] = round(((perc_change['Cases_2023'] - perc_change['Cases_2008']) / perc_change['Cases_2008']) * 100)
```

```python
perc_change['Pct_Change'] = perc_change['Pct_Change'].replace([float('inf'), -float('inf')], pd.NA)
```

```python
perc_change = perc_change.sort_values('Pct_Change', ascending=False).reset_index(drop=True)
```

```python
top10_increase = perc_change.sort_values('Pct_Change', ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top10_increase.sort_values('Pct_Change',ascending=False),
    y='State',
    x='Pct_Change',
    palette='YlGn_r'
)

plt.title('Top 10 States by % Increase in Lyme Disease Cases (20082023)', fontsize=14)
plt.xlabel('Percentage Increase (%)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
davidson = tennessee_counties[tennessee_counties['County'] == 'Davidson County']
```

```python
davidson_long = pd.melt(
    davidson,
    id_vars=['County', 'State'],
    value_vars=[str(year) for year in range(2008, 2024)],
    var_name='Year',
    value_name='Cases'
)
```

```python
davidson_long['Year'] = davidson_long['Year'].astype(int)
davidson_long['Cases'] = pd.to_numeric(davidson_long['Cases'], errors='coerce').fillna(0)
```

```python

plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    data=davidson_long,
    x='Year',
    y='Cases',
    color='#556B2F'
)


for index, row in davidson_long.iterrows():
    barplot.text(
        index, 
        row['Cases']+ .1,  
        int(row['Cases']),
        ha='center',
        fontsize=9,
        color='black'
    )

plt.title('Davidson County Lyme Disease Cases (20082023)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Reported Cases', fontsize=12)
plt.ylim(0, davidson_long['Cases'].max() * 1.1)  
plt.tight_layout()
plt.show()
```
