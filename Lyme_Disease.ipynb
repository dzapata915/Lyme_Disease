```python
county_case =pd.read_csv('data/Case_County_2001-2023.csv')
```

```python
county_case=county_case.rename(columns={'Ctyname':'County','stname':'State'})
```

```python
county_case=county_case.drop(columns=['ststatus','stcode','ctycode'])
```

```python
county_case.columns=['County','State','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']
```

```python
county_case.loc[county_case['State']=='Connecticut']
```

```python
county_case=county_case.dropna()
```

```python
county_case.loc[county_case['State']=='Connecticut']
```

```python
county_case.columns
```

```python
years= ['2001', '2002', '2003', '2004', '2005', '2006',
       '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
county_case['total_cases']=county_case[years].sum(axis=1)
county_case
```

```python
county_case.to_csv('Cleaned County Cases 2001-2023.csv',index=False)
```

```python
count_by_state =pd.read_csv('data/Case Count by State 2008-2023.csv')
```

```python
count_by_state
```

```python
convert_columns=['2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
for col in convert_columns:
    count_by_state[col] = pd.to_numeric(count_by_state[col].str.replace(',', '', regex=False), errors='coerce')
```

```python
count_by_state=count_by_state.drop(index=50)
```

```python
count_by_state[['State','2008']].sort_values(by='2008',ascending=False).head(10)
```

```python
count_by_state[['State','2016']].sort_values(by='2016',ascending=False).head(10)
```

```python
count_by_state[['State','2023']].sort_values(by='2023',ascending=False).head(10)
```

```python
count_by_state['total_cases']=count_by_state[convert_columns].sum(axis=1)
count_by_state
```

```python
count_by_state[['State','total_cases']].sort_values(by='total_cases',ascending=False).head(10)
```

```python
count_by_state['%_change']=round(((count_by_state['2023']-count_by_state['2008'])/count_by_state['2008'])*100,1)
```

```python
count_by_state['State']=count_by_state['State'].str.replace('†','',regex=False)
count_by_state
```

```python
count_by_state[['State','%_change']].sort_values(by='%_change',ascending=False)
```

```python
count_by_state.to_csv('Cleaned State Cases 2008-2023.csv',index=False)
```

```python
Tennessee = count_by_state.loc[count_by_state['State']=='Tennessee']
```

```python
long_TN = Tennessee.melt(
    id_vars=["State", "total_cases", "%_change"],
    var_name="Year",
    value_name="Cases"
)

# Convert year to integer (optional, but useful for sorting in Power BI)
long_TN["Year"] = long_TN["Year"].astype(int)

long_TN= long_TN[['State','Year','Cases']]
long_TN
```

```python
long_TN.plot(kind='bar',x='Year',y='Cases')
```

```python
long_TN.to_csv('TN_state_LD_Casecount.csv',index=False)
```
